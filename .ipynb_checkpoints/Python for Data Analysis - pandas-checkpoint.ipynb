```python
import pandas as pd


web_site = {
    'Day':[1,2,3,4,5,6],
    "Visitor":[100,200,321,343,5334,3534],
    "bounce_rate":[20,10,23,43,32,76]
}

df = pd.DataFrame(web_site)     # to create dataframes from dictionary using pandas


# Slicing
print (df.head(2))   # slicing with head() 
print (df.tail(2))   # slicing with tail()


# merging

df1 = pd.DataFrame({
    "HPI":[80,90,70,60],
    "Int_rate":[2,3,4,5],
    "IND_GDP":[45,56,67,78]
}, index = [2001,2002,2003,2004])

df2 = pd.DataFrame({
    "HPI":[80,90,70,60],
    "Int_rate":[2,3,4,5],
    "IND_GDP":[45,56,67,78]
}, index = [2005,2006,2007,2008])

merge = pd.merge(df1,df2, on = "HPI")
print (merge)



# Joining

df3 = pd.DataFrame({
    "employment_rate":[2,3,4,5],
    "IND_Curr":[45,56,67,78]
}, index = [2001,2009,2007,2008]) 

joined = df3.join(df2)
print (joined)


#     Changing the column headers and index

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")

df4 = pd.DataFrame({
    "Day": [1,2,3,4], 
    "visitors":[200,300,40,500], 
    "bounce_rate":[20,340,45,56]
})
 
df4.set_index("Day", inplace =True)  # changing index
#print (df4) 

df4.plot()        # to plot the graph using matplotlib
#plt.show()

df4 = df4.rename(columns={"visitors":"users"}) # changing column headers
print (df4)


# concatanation

concat = pd.concat([df1,df2])
print (concat)

# data munching 

data = pd.read_csv("/home/shantanu/Documents/Desktop/Document/AWS-CDA/credentials.csv") # to read .csv
data.to_html("data_in.html") # to convert .csv to html


# to analyze population data

population = pd.read_csv("/home/shantanu/Documents/Python3/python3/population_data.csv")

dfp = population.head(5)
dfp = dfp.set_index(["Country Name"])

sd = dfp.reindex(columns=['1970','1971'])

p = sd.diff(axis=0)
p.plot(kind = 'bar')
plt.show()


# statistics

from statistics import mean

print (mean([1,2,3,4,43,2,11,2]))

from statistics import mode

print (mode([1,2,3,4,43,2,11,2]))

from statistics import median

print (median([1,2,3,4,43,2,11,2]))

from statistics import variance

print (variance([1,2,3,4,43,2,11,2]))
```
