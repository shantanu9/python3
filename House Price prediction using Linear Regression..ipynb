### Prediction of price of a house based on given housing dataset using Machine Learning.

Output of linear regression is quantitative in nature and directly proportional to the variables. 

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston
```

```python
df_x = pd.DataFrame(boston.data,columns=boston.feature_names)
df_y = pd.DataFrame(boston.target)
df_x, df_y
```

```python
reg = linear_model.LinearRegression()   # we are using linear regression model for training our data
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_x,df_y,test_size=0.2,random_state=4)  # test_size .20 means that 20 % of data will be used in testing, rest will be for training.
```

```python
reg.fit(x_train,y_train)        # training the regression model for train input and labelled output.
```

```python
reg.coef_
```

```python
a= reg.predict(x_test)   # prediction means to get the output label for input label once the model is trained.
a
```

```python
np.mean((a-y_test)**2)   # mean sqaured value.
```
