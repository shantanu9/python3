### Machine Learning Workflow

Define Objective > Prepare Data > Collect Data > Select Algorithm >  Train Model > Test Model > Predict. 

### Types of Machine Learning

1. Supervised : ML model learns from the past input data and makes future prediction as output. 
2. Unsupervised : ML model uses unlabeled input data and allow the algorithm to act on that information without guidance. Trying to find the hidden structure/pattern in unlabelled data.
3. Reinforcement:  learning involves teaching the machine to think for itself based on it's past action reward. Algos:  Q-Learning,  Temporal difference(TD), Deep Adversarial Networks

### Techniques in Supervised
1.  Classification:  is concerned with building models that separate data into distinct classes.  

Algos:  Decision Tree, Support Vector machine,  Logistics regression

2. Regression:  Based on the previous input data, the machine predicts continuous output value. 

Algos: Linear Regression,  Polynomial Regression 

Linear Regression Demo:  Housing example

```python
# importing Numpy, Matplotlib, sklearn libraries

import matplotlib.pyplot as plt
import numpy as np

# datasets from scikit learn

from sklearn import datasets, linear_model
```

```python
# Load the datasets
house_price = [245,312,279,308,199,219,405,324,319,255]
size = [1400,1600,1700,1875,1100,1550,2350,2450,1425,1700]
house_price, size

# Reshape the input to your regression
size2 = np.array(size).reshape((-1,1))
size2
```

```python
# By using fit module in Linear Regression, user can fit the data frequently and quickly.

reg = linear_model.LinearRegression()
reg.fit(size2,house_price)
print("coefficients: ", reg.coef_)
print("intercept: ", reg.intercept_)
```

```python
# prediction

size_new = 1400
print("Price of the new size: ",reg.predict([[size_new]]))
```

```python
# formula obtained fro the trained model
def graphh(formula, x_range):
    x = np.array(x_range)
    y = eval(formula)
    plt.plot(x,y)

graphh('reg.coef_*x + reg.intercept_', range(1000,3700))
plt.scatter(size, house_price, color = 'black')
plt.ylabel('house price')
plt.xlabel('size of house')
plt.show()
```

#### Techniques in unsupervised. 
1. Clustering:  is used analyzing and grouping data which does not include pre-labeled class or even a class attribute at all.

Cluster is a collection of objects which are "similar" between them and are "dissimilar" to the objects belongings to other clusters. 

algos:  K-means,  Hierarchical Clustering,  Hidden Markov model.

2. Association:  Discovers the probability oof the co-occurence of items in a collection

algos:  Apriori, FP-Growth.

###### K-means Clustering Demo

Step 1: First, we need to randomly initialize 2 points called the cluster centroids. 

Step 2: Now, based upon the distance from the orange cluster centroid or green cluster centroid, it will group itself into that particular group. 

Step 3: Move Centroids - Now, you will take the two cluster centroids and iteratively reposition them for optimization.  

Step 4: Repeat previous 2 steps iteratively till the cluster centroids stop changing their positions and become static.

step 5: Once the clusters become static then k-means clustering algorithm is said to be converged. 

```python
# importing Numpy, Matplotlib, and sklearn libraries. 

import numpy as np
import matplotlib.pyplot as plt

from matplotlib import style
style.use("ggplot")
from sklearn.cluster import KMeans
```

```python
# Plotting and visualizing our data before feeding it into k-means algo
x = [1,5,1.5,8,1,9]
y = [2,8,1.8,8,0.6,11]
plt.scatter(x,y)
plt.show()
```

```python
# Converting our data to a NumPy array

X = np.array([[1,2],[5,8],[1.5,1.8],[8,8],[1,0.6],[9,11]])

# initialize K-means algorithm with the required parameters and we fit the data
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
# getting the values of centroid and the labels based on the fit. 
# kmeans.labels_ will put the input(x,y) into clusters

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print (centroids)
print (labels)

# 0 indicates first cluster and 1 is the second cluster
```

```python
# Plotting and visualizing output

colors = ["g.","r.","c.","y."]

for i in range(len(X)):
    print("coordinate:", X[i], "label:", labels[i])
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)
    
plt.scatter(centroids[:,0], centroids[:,1], marker = "x", s=150, linewidths = 5, zorder = 10)
plt.show()
```

```python
# initialize K-means algorithm with the required parameters and we fit the data
kmeans_weather = KMeans(n_clusters=3)
kmeans_weather.fit(X)
```

```python
# Plotting and visualizing output

colors = ["g.","b.","y.","c."]

for i in range(len(X)):
    print("coordinate:", X[i], "label:", labels[i])
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)
    
plt.scatter(centroids[:,0], centroids[:,1], marker = "x", s=150, linewidths = 5, zorder = 10)
plt.show()
```
