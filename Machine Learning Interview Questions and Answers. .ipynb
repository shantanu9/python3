What is overfitting? How can you avoid it?

Overfitting occurs when the model learns the training set too well.  It takes up random fluctuations in the training data as concepts.  This impact the model's ability to generalize and don't apply to new data. 

High Loss and low accuracy is seen in the test dataset. 

3 main methods to avoid overfitting:
Regularization: involves a cost term for the features involved with the objective function.
Simple model: with lesser variables and parameters, the variance can be reduced. 
Cross-validation methods, like k-folds



What is 'Training Set' and 'Test Set' in ML model ?

Training set are given to model to analyze and learn. usually 70% is taken as Training set. This is labelled data used to train the model.  

Test set is used to test the accuracy of the hypothesis generated by the model. Remaining 30% is taken as Testing dataset.  We test without labeled data and then verify results with labels. 

How do you handle missing or corrupted data in a dataset ?

To drop rows/colums or replace them completely with some other values

2 useful methods in pandas:  

isnull() and dropna() will help finding the columns/rows with the missing data and drop them.

Fillna() will replace the wrong values with a placeholder value(0)
